```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image1 = cv.imread('images/obama.jpg', -1)
image2 = cv.imread('images/obama.jpg', 0)
image3 = cv.imread('images/obama.jpg', 1)
```

```python
cv.imshow('Image 1', image1)
cv.waitKey(0)
cv.imshow('Image 2', image2)
cv.waitKey(0)
cv.imshow('Image 3', image3)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img = image1
cv.namedWindow('Windows image', cv.WINDOW_NORMAL)
cv.imshow('Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imwrite("new.png", img)
```
